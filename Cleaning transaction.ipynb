```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df_transactions=pd.read_excel('KPMG_VI_New_raw_data_update_final.xlsx',sheet_name='Transactions')
```

```python
df_transactions.head()
```

```python
df_transactions.dtypes
```

```python
df_transactions['transaction_id'].isnull().sum()
```

```python
df_transactions['transaction_id'].nunique()
```

```python
#it is better to drop transaction id as it doesnt cause any influene
df_transactions.drop(['transaction_id'],axis=1,inplace=True)
```

```python
df_transactions.drop('product_id',axis=1,inplace=True)
```

```python
df_transactions.drop('customer_id',axis=1,inplace=True)
```

```python
from datetime import datetime
df_transactions['year']=df_transactions['transaction_date'].apply(lambda date:date.year)
```

```python
df_transactions['month']=df_transactions['transaction_date'].apply(lambda date:date.month)
df_transactions['day']=df_transactions['transaction_date'].apply(lambda date:date.day)
```

```python
df_transactions.drop('transaction_date',axis=1,inplace=True)
```

```python
df_transactions.dtypes
```

```python
df_transactions['online_order'].isnull().sum()
```

```python
df_transactions['online_order'].mode()
```

```python
df_transactions['online_order'].unique()
```

```python
df_transactions['online_order'].replace(np.nan,1.0,inplace=True)
```

```python
df_transactions['online_order'].isnull().sum()
```

```python
df_transactions['order_status'].unique()
```

```python
df_transactions['order_status']=df_transactions['order_status'].map({'Approved':int(1),'Cancelled':int(0)})
```

```python
df_transactions['online_order'].unique()
```

```python
df_transactions['brand'].unique()
```

```python
brand_dummy=pd.get_dummies(df_transactions['brand'],drop_first=True)
```

```python
brand_dummy
```

```python
df_transactions=pd.concat([df_transactions,brand_dummy],axis=1)
```

```python
df_transactions.drop('brand',axis=1,inplace=True)
```

```python
df_transactions['product_line'].replace(np.nan,'Standard',inplace=True)
```

```python
product_type=pd.get_dummies(df_transactions['product_line'],drop_first=True)
```

```python
df_transactions=pd.concat([df_transactions,product_type],axis=1)
```

```python
df_transactions.drop('product_line',axis=1,inplace=True)
```

```python
df_transactions
```

```python
df_transactions['product_class'].mode()
```

```python
df_transactions.replace(np.nan,'medium',inplace=True)
```

```python
df_transactions['product_class'].unique()
```

```python
class_dummy=pd.get_dummies('product_class',drop_first=True)
```

```python
df_transactions=pd.concat([df_transactions,class_dummy],axis=1)
```

```python
df_transactions.drop('product_class',axis=1,inplace=True)
```

```python
df_transactions
```

```python
df_transactions['product_size']=df_transactions['product_size'].map({'medium':int(1), 'large':int(2), 'small':int(0)})
```

```python
df_transactions
```

```python
df_transactions['list_price']=df_transactions['list_price'].apply(np.round)
```

```python
df_transactions
```

```python
df_transactions['standard_cost'].isnull().sum()
```

```python
df_transactions['product_first_sold_date']
```
